```python

import matplotlib.pyplot as plt
import numpy as np
r=0.06

def pe_n(g,n):
  return np.power((1+g)/(1+r),n)/r
  


g = np.linspace(0,5*r,500)
pe_lin = g/r/r

year_list=[1,2,3,4,5,10,15]
pe_year=[]
for year in year_list:
    pe_year.append(pe_n(g,year))

plt.rcParams['figure.figsize'] = (20.0, 10.0) # 
plt.plot(g,pe_lin,color = 'black',linestyle='--',label = 'pe_lin')         #labelplot
for i,year in enumerate(year_list):
    plt.plot(g,pe_year[i],label="pe_"+str(year))
plt.legend(loc='best')
 
# plt.savefig('.//result//3.7.png')
```
