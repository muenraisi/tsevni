```python

import matplotlib.pyplot as plt
import numpy as np
r=0.05

def pe_n(g,n):
  return np.power((1+g)/(1+r),n)/r
  


g = np.linspace(0,5*r,500)
pe_lin = g/r/r

year_list=[1,2,3,4,5,10,15,20,50]
pe_year=[]
for year in year_list:
    pe_year.append(pe_n(g,year))
 
plt.plot(g,pe_lin,color = 'r',label = 'pe_lin')         #label每个plot指定一个字符串标签
for i in range(len(year_list)):
    plt.plot(pe_year[i])
plt.legend(loc='best')
 
# plt.savefig('.//result//3.7.png')
```
